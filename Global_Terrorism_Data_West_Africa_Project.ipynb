```python
# importing dependencies
import requests
from scrapy import Selector
from scrapy.crawler import CrawlerProcess
import pandas as pd
```

```python
## getting the data from the GTD website into html format with the 'get' method of request
html = requests.get('https://www.start.umd.edu/gtd/search/Results.aspx?page=3&search=Nigeria&count=100&charttype=line&chart=overtime&expanded=yes&ob=CountryText&od=asc#results-table').content
```

```python
#xpath that leads to the table
xpath_to_table = '//tbody/tr'
# xpath that leads to inside the table
in_table = '//td/text()'
```

```python
# creating select object with Selector from scrapy 
sel = Selector(text = html)
```

```python
tab = sel.xpath(xpath_to_table)
tab
```

```python
# for loop that extracts the elements in the table to to create a dictionary
for row in tab:
   table_dic = {
        'Date': table.xpath('//td[2]/text()').extract(),
       'Country': table.xpath('//td[3]/text()').extract(),
       'City': table.xpath('//td[4]/text()').extract(),
       'Perpetraotor': table.xpath('//td[5]/text()').extract(),
       'Fatalities': table.xpath('//td[6]/text()').extract(),
       'Injured': table.xpath('//td[7]/text()').extract(),
       'Target': table.xpath('//td[8]/text()').extract(),
       'Region': table.xpath('//td[9]/text()').extract(),
       'Attack_type': table.xpath('//td[10]/text()').extract(),
       'Weapon_type':table.xpath('//td[11]/text()').extract()
       
        
    }
```

```python
# converting dictionary to dataframes using pandas
table = pd.DataFrame(table_dic)

```
